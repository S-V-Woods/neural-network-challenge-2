## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras import layers

from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
from pandas.api.types import is_string_dtype
from pandas.api.types import is_numeric_dtype

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[["Attrition","Department"]]
y_df.head()
```

```python
# Create a list of at least 10 column names to use as X data

x = [
"Education",
"Age",
"DistanceFromHome",
"JobSatisfaction",
"OverTime",
"StockOptionLevel",
"WorkLifeBalance",
"YearsAtCompany",
"YearsSinceLastPromotion",
"NumCompaniesWorked"]
#print(x)

# Create X_df using your selected columns
X_df = attrition_df[x]

# Show the data types for X_df
X_df.dtypes
```

```python
X_df["OverTime"] = pd.get_dummies(X_df["OverTime"], drop_first = True, dtype = int).values.reshape(-1,1)
```

```python
# encoder = OrdinalEncoder()
# X_df["OverTime"] = encoder.fit_transform(X_df[["OverTime"]])
```

```python
# #is_non_numeric = pd.to_numeric(df['column'], errors='coerce').isnull()
# encoder = OrdinalEncoder(categories=[['No', 'Yes']])
# ['No', 'Yes']]X_df["OverTime"] = encoder.fit_transform(X_df[["OverTime"]])
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X = X_df
y = y_df
# for i in range(38): #Find the random state that matches the starter code
#     X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=i)
#     print(X_train["OverTime"].value_counts())
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=37)
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
encoder = OrdinalEncoder()
encoder.fit_transform(X_train[["OverTime"]])
X_train[["OverTime"]] = encoder.fit_transform(X_train[["OverTime"]])
X_test[["OverTime"]] = encoder.fit_transform(X_test[["OverTime"]])

print(X_train[["OverTime"]].value_counts())
print(X_test[["OverTime"]].value_counts())
display(X_train)
display(X_test)
```

```python
# Create a StandardScaler
scaler = StandardScaler()

# Fit the StandardScaler to the training data
scaler = scaler.fit(X_train)

# Scale the training and testing data
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
y_train
```

```python
# Create a OneHotEncoder for the Department column
ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False, dtype='int')

# Fit the encoder to the training data
ohe.fit(y_train[["Department"]])

# Create two new variables by applying the encoder
# to the training and testing data
y_train_dep_encoded = ohe.transform(y_train[["Department"]])
y_test_dep_encoded = ohe.transform(y_test[["Department"]])

# ohe.fit_transform(y_train[["Department"]])
# ohe.fit_transform(y_train[["Department"]])
print([y_train_dep_encoded])
print([y_test_dep_encoded])
```

```python
# Create a OneHotEncoder for the Attrition column
ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False, dtype='int')

# Fit the encoder to the training data
ohe.fit(y_train[["Attrition"]])

# Create two new variables by applying the encoder
# to the training and testing data
y_train_att_encoded = ohe.transform(y_train[["Attrition"]])
y_test_att_encoded = ohe.transform(y_test[["Attrition"]])

# ohe.fit_transform(y_train[["Department"]])
# ohe.fit_transform(y_train[["Department"]])
print([y_train_att_encoded])
print([y_test_att_encoded])
```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
#X_train.columns
display(len(X_train.columns))
# Create the input layer

input_layer = layers.Input(shape=(X_train.shape[1],), name='input_features')

# Create at least two shared layers
shared_layer1 = layers.Dense(64, activation='relu', name = "shared_layer1")(input_layer)
shared_layer2 = layers.Dense(128, activation='relu', name = "shared_layer2")(shared_layer1)
display(X_train.columns)


```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer


# Create the output layer
#department_output = layers.Dense(3, activation='softmax', name='department_output')(shared_layer2)

# pose_dense = layers.Dense(64, activation='relu')(dense_shared)
# pose_output = layers.Dense(len(y_train_pose.columns),
#                            activation='softmax',
#                              name='pose_output')(pose_dense)
#shared_layer2 = layers.Dense(128, activation='relu')(shared_layer1)
department_hidden = layers.Dense(32, activation='relu',name='department_hidden')(shared_layer2)
department_output = layers.Dense(y_train_dep_encoded.shape[1],activation='softmax',name='department_output')(department_hidden)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer


# Create the output layer
# attrition_output = layers.Dense(3, activation='softmax', name='attrition_output')(shared_layer2)
attrition_hidden = layers.Dense(32, activation='relu', name= "attrition_hidden")(shared_layer2)
# attrition_output = layers.Dense(y_train_att_encoded.shape[1],activation='sigmoid',name='attrition_output')(department_dense)
attrition_output = layers.Dense(y_train_att_encoded.shape[1],activation='sigmoid',name='attrition_output')(attrition_hidden)
```

```python
# Create the model
model = Model(inputs=input_layer, outputs=[
    department_output,
    attrition_output
])

# Compile the model
model.compile(optimizer='adam',
            # loss={'department_output': 'categorical_crossentropy',
            #       'attrition_output': 'binary_crossentropy'
            #      },
            loss={'department_output': 'categorical_crossentropy',
                  'attrition_output': 'categorical_crossentropy'
                 },
            # loss={'department_output': tf.keras.losses.CategoricalCrossentropy(from_logits=False),
            # 'attrition_output': tf.keras.losses.CategoricalCrossentropy(from_logits=False)
            # },
            metrics={'department_output': 'accuracy',
            'attrition_output': 'accuracy'
            })

# Summarize the model
model.summary()
```

```python
# Train the model

model.fit(
    X_train,
    {
        'department_output': y_train_dep_encoded,
        'attrition_output': y_train_att_encoded
        # 'department_output': department_output,
        # 'attrition_output': attrition_output
        #'department_output': y_df["Department"],
        #'attrition_output': y_df["Attrition"]
    },
    epochs=100,  # You can adjust the number of epochs based on your needs
    batch_size=32,  # You can adjust the batch size based on your available memory
    #validation_split=0.2  # You can specify the validation split if you have a separate validation set
)
```

```python
# Evaluate the model with the testing data
test_results = model.evaluate(X_test, {'department_output': y_train_dep_encoded, 'attrition_output': y_train_att_encoded})
test_results
```
